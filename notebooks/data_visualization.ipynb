# Visualização dos dados utilizando matplotlib e seaborn

Como o unico dado do dataset são textos, a visualização fica baseada em conceitos básicos como contagem de palavras por gênero e quantidade de palavras por música

## Antes de iniciar com as vizualizações, é necessário fazer a limpeza básica dos textos

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10.0, 6.0)
import plotly.graph_objs as go
pd.options.display.max_columns = 30
pd.options.plotting.backend = "plotly"
from plotly.offline import iplot 
```

```python
def clean_text(text):
    text = text.str.replace('\n',' ')
    text = text.str.replace('  ',' ')
    text = text.str.lower()
    text = text.str.replace(',','')
    text = text.str.replace('?','')
    text = text.str.replace('!','')
    text = text.str.replace('.','')
    return text
```

```python
ds['lyric'] = clean_text(ds['lyric'])
```

```python
ds['lyric']
```

```python
ds['word_count'] = ds['lyric'].apply(lambda x: len(str(x).split()))
```

```python
ds['len'] = ds['lyric'].str.len()
```

```python
ds
```

```python
x1 = ds.loc[ds['genero'] == 1, 'len']
x2 = ds.loc[ds['genero'] == 2, 'len']
x3 = ds.loc[ds['genero'] == 3, 'len']
x4 = ds.loc[ds['genero'] == 4, 'len']

trace1 = go.Histogram(
    x=x1, name='Bossa Nova',
    opacity=0.75
)
trace2 = go.Histogram(
    x=x2, name = 'Funk',
    opacity=0.75
)
trace3 = go.Histogram(
    x=x3, name = 'Gospel',
    opacity=0.75
)
trace4 = go.Histogram(
    x=x4, name = 'Sertanejo',
    opacity=0.75
)

data = [trace1, trace2, trace3, trace4]
layout = go.Layout(barmode='overlay', title='Distribuição do tamanho das letras por gêneros')
fig = go.Figure(data=data, layout=layout)

fig.show()
```

```python
x1 = ds.loc[ds['genero'] == 1, 'word_count']
x2 = ds.loc[ds['genero'] == 2, 'word_count']
x3 = ds.loc[ds['genero'] == 3, 'word_count']
x4 = ds.loc[ds['genero'] == 4, 'word_count']

trace1 = go.Histogram(
    x=x1, name='Bossa Nova',
    opacity=0.75
)
trace2 = go.Histogram(
    x=x2, name = 'Funk',
    opacity=0.75
)
trace3 = go.Histogram(
    x=x3, name = 'Gospel',
    opacity=0.75
)
trace4 = go.Histogram(
    x=x4, name = 'Sertanejo',
    opacity=0.75
)

data = [trace1, trace2, trace3, trace4]
layout = go.Layout(barmode='overlay', title='Distribuição do numero de palavras por gêneros')
fig = go.Figure(data=data, layout=layout)

fig.show()
```
